Cerco di visualizzare una normale standardizzata

Importo dunque le lib che mi servono

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

Assegno a z una normale standardizzata

```python
Z = stats.norm()
```

```python
x = np.arange(-5,5, 0.1)
```

```python
%matplotlib inline
plt.plot(x, Z.pdf(x))
plt.show()
```

```python
%matplotlib inline
plt.plot(x,Z.cdf(x))
plt.show()
```
